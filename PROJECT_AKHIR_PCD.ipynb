```python
# =====================================================
# DETEKSI AKTIVITAS MENCURIGAKAN PADA VIDEO CCTV
# MENGGUNAKAN YOLOv8
# Google Colab Version
# =====================================================

# Install Library
!pip install -q ultralytics opencv-python

import cv2
import os
from ultralytics import YOLO
from google.colab import files
from IPython.display import Video

# =====================================================
# Upload Video
# =====================================================

print("===================================")
print("UPLOAD VIDEO CCTV")
print("===================================")

uploaded = files.upload()

video_path = list(uploaded.keys())[0]

print("Video berhasil dipilih :", video_path)

# =====================================================
# Load Model YOLOv8
# =====================================================

print("\nLoading YOLOv8 Model...")

model = YOLO("yolov8n.pt")

print("Model berhasil dimuat!")

# =====================================================
# Membaca Video
# =====================================================

cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    raise Exception("Video tidak dapat dibuka!")

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

fps = cap.get(cv2.CAP_PROP_FPS)

# Jika FPS terbaca 0
if fps == 0:
    fps = 30

print(f"Resolusi : {width} x {height}")
print(f"FPS      : {fps}")

# =====================================================
# Output Video
# =====================================================

output_video = "hasil_deteksi.mp4"

fourcc = cv2.VideoWriter_fourcc(*'mp4v')

out = cv2.VideoWriter(
    output_video,
    fourcc,
    fps,
    (width, height)
)

if not out.isOpened():
    raise Exception("VideoWriter gagal dibuat!")

print("Output video berhasil dibuat!")

# =====================================================
# Mulai Deteksi
# =====================================================

frame_number = 0

print("\nMemproses video...\n")

while True:

    ret, frame = cap.read()

    if not ret:
        break

    frame_number += 1

    # Prediksi YOLO
    results = model.predict(
        frame,
        conf=0.4,
        verbose=False
    )

    person_count = 0

    # ==========================
    # Bounding Box
    # ==========================

    for result in results:

        for box in result.boxes:

            cls = int(box.cls[0])

            if cls == 0:

                person_count += 1

                x1, y1, x2, y2 = map(int, box.xyxy[0])

                confidence = float(box.conf[0])

                cv2.rectangle(
                    frame,
                    (x1, y1),
                    (x2, y2),
                    (0,255,0),
                    2
                )

                cv2.putText(
                    frame,
                    f"Person {confidence:.2f}",
                    (x1,y1-10),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.6,
                    (0,255,0),
                    2
                )

    # =================================================
    # Deteksi Aktivitas Mencurigakan
    # =================================================

    status = "NORMAL"

    warna = (0,255,0)

    if person_count >= 5:

        status = "SUSPICIOUS ACTIVITY"

        warna = (0,0,255)

    cv2.putText(
        frame,
        f"Status : {status}",
        (20,40),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        warna,
        3
    )

    cv2.putText(
        frame,
        f"Persons : {person_count}",
        (20,85),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (255,0,0),
        3
    )

    # Simpan frame
    out.write(frame)

    # Progress
    if frame_number % 50 == 0:
        print(f"Frame diproses : {frame_number}")

# =====================================================
# Selesai
# =====================================================

cap.release()
out.release()

print("\n====================================")
print("Deteksi selesai!")
print("====================================")

# =====================================================
# Konversi ke H264 supaya bisa diputar
# =====================================================

print("Mengkonversi video...")

!ffmpeg -loglevel error -y -i hasil_deteksi.mp4 -vcodec libx264 hasil_final.mp4

print("Konversi selesai!")

# =====================================================
# Tampilkan Video
# =====================================================

print("\nMenampilkan hasil...\n")

Video("hasil_final.mp4", embed=True)

# =====================================================
# Download
# =====================================================

files.download("hasil_final.mp4")
```
